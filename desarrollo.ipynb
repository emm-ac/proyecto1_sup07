Cambiar el tipo de la columna date_added a datetime

```python
#datos['date_added'] = pd.to_datetime(datos['date_added'], format='%m/%d/%Y')
#datos['date_added'] = pd.to_datetime(datos['date_added'])
datos['date_added'] = pd.to_datetime(datos['date_added'], errors='coerce')
```

Ordenar de mas antiguo a mas reciente usando : ".sort_values(by=['date_added'])"

```python
datos.sort_values(by=['date_added'], inplace=True)
```

Reemplazar en la columna "director" el valor "Not Given" por None

```python
datos['director'] = datos['director'].replace('Not Given',None)
```

Usando mascaras dividir en 3 dataframes distintos, "release_year" 2019, 2020, 2021

```python
mask_2019 = datos["release_year"].str.contains("2019", na=False)
mask_2020 = datos["release_year"].str.contains("2020", na=False)
mask_2021 = datos["release_year"].str.contains("2021", na=False)
```

```python
datos_2019 = datos[mask_2019]
datos_2020 = datos[mask_2020] 
datos_2021 = datos[mask_2021]
```

Transformar los 3 dataframes en 3 diccionarios distintos usando: .reset_index().to_dict(orient="index")

```python
datos_2019.reset_index().to_dict(orient="index")
datos_2020.reset_index().to_dict(orient="index")
datos_2021.reset_index().to_dict(orient="index")
```
